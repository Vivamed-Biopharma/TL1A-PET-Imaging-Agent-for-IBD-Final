```python
import re
AA = set("ACDEFGHIKLMNPQRSTVWY")

def qc(seq):
    issues=[]
    if any(ch not in AA for ch in seq): issues.append("non-AA")
    if re.search(r"\\s", seq): issues.append("whitespace")
    if re.search(r"N[^P][ST]", seq): issues.append("NXS/T glyco motif")
    return issues

bad=[]
for name, ch in FABS.items():
    for chain in ("VH","VL"):
        issues = qc(ch[chain])
        if issues: bad.append((name, chain, issues))

if bad:
    for b in bad: print("QC FAIL:", b)
    raise SystemExit("Fix sequences above and rerun.")
else:
    print("QC PASS: no illegal characters; no whitespace; no NXS/T motifs.")

# Print FASTA for each chain
for name,ch in FABS.items():
    print(f">{name}_VH\n{ch['VH']}\n>{name}_VL\n{ch['VL']}\n")
```
